```python
out_liers=['Pregnancies','BloodPressure','Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
for col in out_liers:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  ul=q3+1.5*iqr
  ll=q1-1.5*iqr
  df=df[(df[col]>=ll)&(df[col]<=ul)]
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('Outcome',axis=1)  # axis=1 tells pandas to drop a column entire, not an index
y = df["Outcome"]
x
```

🎯 Use of GridSearchCV

👉 The main use of GridSearchCV is to automatically find the best parameters (hyperparameters) for your machine learning model — in your case, the Random Forest Classifier.

🧠 Think of it like this:

When you build a model (like RandomForest), you have many settings you can change:

How many trees to grow (n_estimators)

How deep the trees can go (max_depth)

How many samples to split (min_samples_split)

etc.

These are called hyperparameters — and they greatly affect your model’s accuracy.
But you don’t know which combination works best.

So instead of manually guessing 👇

rf = RandomForestClassifier(n_estimators=100, max_depth=10)


you let GridSearchCV try many combinations automatically and pick the best one.

⚙️ What GridSearchCV does:

Takes all parameter combinations from the grid (you defined in param_grid).

For each combination:

It trains your model multiple times (cross-validation).

It tests how well the model performs.

Finds the combination that gives the highest accuracy (or any metric you choose).

📊 Example

Let’s say you test:

n_estimators	max_depth	Accuracy
100	10	0.78
200	20	0.82
500	30	0.80

GridSearchCV will automatically select:

Best Parameters: {'n_estimators': 200, 'max_depth': 20}
Best Accuracy: 0.82


Now you can build your final model with these best settings to get maximum accuracy and stability.

💡 So in simple words:
Without GridSearchCV	With GridSearchCV
You guess the parameters manually	It tests all combinations for you
Risk of underfitting or overfitting	Finds best balance automatically
One model trained	Many models tested
Lower accuracy	Optimized accuracy

✅ Final Use:

To automatically tune and select the best hyperparameters that make your model perform better and more reliable on unseen data.

gradient boosting


```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
gb = GradientBoostingClassifier(
    n_estimators=300,       # number of boosting stages
    learning_rate=0.1,      # step size shrinkage
    max_depth=2,            # depth of individual estimators
    random_state=42
)

# Train model
gb.fit(x_train, y_train)

# Predict on test data
y_pred = gb.predict(x_test)

# Evaluate performance
acc = accuracy_score(y_test, y_pred)
print("✅ Accuracy:", acc)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

1. Pregnancies

How many times a woman has been pregnant.

More pregnancies can slightly increase diabetes risk.

2. Glucose

Sugar level in the blood.

Higher sugar → higher chance of diabetes. This is the most important factor.

3. BloodPressure

The force of blood in the arteries.

Higher blood pressure can be linked to diabetes risk.

4. SkinThickness

How thick the fat under the skin is.

Higher fat → higher risk.

5. Insulin

Amount of insulin in the blood.

Very high or low → indicates problems controlling sugar.

6. BMI

Body Mass Index — how heavy someone is for their height.

Higher BMI → higher diabetes risk.

7. DiabetesPedigreeFunction

Shows if diabetes runs in the family.

Higher value → higher genetic risk.

8. Age

Older people are more likely to get diabetes.

9. Outcome

0 = No Diabetes, 1 = Diabetes (this is what we’re predicting).

General Inferences from the Dataset

Glucose is the strongest predictor — high glucose almost always leads to diabetes.

BMI, Age, Insulin, and DiabetesPedigreeFunction are secondary predictors — they refine the risk assessment.

Pregnancies and BloodPressure have smaller but still notable effects.

Features interact: For example, a patient with moderate BMI but high glucose and strong family history may still be predicted as diabetic.

Imbalanced classes: Usually, fewer patients have diabetes than don’t — this affects precision/recall.
High BMI + high glucose → strong risk → model predicts 1.
High-risk patients: Glucose > 125, BMI > 30, Age > 40, strong family history → Outcome = 1

Low-risk patients: Glucose < 100, BMI < 28, younger age, low family history → Outcome = 0
